```python
# make sure that your image consists of only 3 colours, 
# the colour of the text (usually black), the colour of background (usually white), 
# and the colour of watermarks (usually gray), but can be modified easily for images with multiple colors, 
# and different file formats like pdfs
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img_arr = cv2.imread("input_img.jpg")  # replace with your image path
img_arr = cv2.cvtColor(img_arr, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_arr)
plt.show()
```

```python
X = np.reshape(img_arr, (-1, 3))
```

```python
# find the most dominant 3 colours in the image

from sklearn.cluster import KMeans
clf = KMeans(n_clusters=3)
clf.fit(X)
clf.cluster_centers_
```

```python
y_pred = clf.predict(X)
```

```python
for i, color in enumerate(clf.cluster_centers_):
    plt.subplot(3, 3, i+1)
    plt.axis("off")
    plt.title(i)
    
    color_img = np.zeros((100, 100, 3), dtype=np.uint8)
    color_img[:, :, :] = color
    
    plt.imshow(color_img)

plt.show()
```

```python
# See the index corresponding to the colours of input image from the above swatch
# then, assign each index to the following fields

bg_index = 0  # index of color of background
txt_index = 1  # index of color of text
watermark_index = 2  # index of color of watermarks
```

```python
# replace each pixel colored as watermark, with the color of background
for i in range (len(X)):
    index = y_pred[i]
    if (index == watermark_index):
        index = bg_index
    color = clf.cluster_centers_[index]
    X[i] = color
    
```

```python
X_ = np.copy(X)
X_ = np.array([X_])
X_ = X_.reshape((img_arr.shape[0], img_arr.shape[1], 3))

# to bolden the text a lil bit
for i in range(img_arr.shape[0]):
    for j in range(img_arr.shape[1]):
        comparison = X_[i][j] == clf.cluster_centers_[1]
        if comparison.all():
            X_[i-1][j] = clf.cluster_centers_[1]
            X_[i][j-1] = clf.cluster_centers_[1]
            
```

```python
plt.imshow(X_)
plt.show()
```

```python
cv2.imwrite("output_img.jpg", X_)
```
